```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Selenium 드라이버 초기화
driver = webdriver.Chrome()
driver.get('https://transportation.asamaru.net/%EC%8B%9C%EC%99%B8%EB%B2%84%EC%8A%A4/%EC%8B%9C%EA%B0%84%ED%91%9C/%EC%B2%AD%EC%A3%BC%EC%8B%9C%EC%99%B8%EB%B2%84%EC%8A%A4%ED%84%B0%EB%AF%B8%EB%84%90/')  # 실제 URL로 변경

# 터미널 정보 가져오기
terminals = driver.find_elements(By.CSS_SELECTOR, '.cTimetable')

terminal_data = {}

# 각 터미널의 이름과 출발 시간 수집
for terminal in terminals:
    terminal_name = terminal.find_element(By.CSS_SELECTOR, '.toc').text  # 터미널 이름
    departure_times = terminal.find_elements(By.CSS_SELECTOR, '.departure')  # 출발 시간 목록
    prices= terminal.find_elements(By.CSS_SELECTOR, '.fare')
    duration_times = terminal.find_elements(By.CSS_SELECTOR, '.duration')
    # 출발 시간 텍스트를 리스트로 저장
    times = [time.text for time in departure_times]
    fares = [price.text for price in prices]
    durations = [duration.text for duration in duration_times]
    terminal_data[terminal_name] = {
        'times': times,
        'fares': fares,
        'durations': durations
   }

# 수집한 데이터 출력
for terminal_name, data in terminal_data.items():
    print(f"터미널: {terminal_name}")
    for time, fare, duration in zip(data['times'], data['fares'], data['durations']):
        print(f"{time},{fare},{duration}")

# 드라이버 종료
driver.quit()
```

```python

import csv

# Selenium 드라이버 초기화
driver = webdriver.Chrome()

# 웹 페이지 접속
driver.get('https://transportation.asamaru.net/%EC%8B%9C%EC%99%B8%EB%B2%84%EC%8A%A4/%EC%8B%9C%EA%B0%84%ED%91%9C/%EC%B2%AD%EC%A3%BC%EC%8B%9C%EC%99%B8%EB%B2%84%EC%8A%A4%ED%84%B0%EB%AF%B8%EB%84%90/')

# 터미널 정보 가져오기
terminals = driver.find_elements(By.CSS_SELECTOR, '.cTimetable')
""""# CSV 파일을 쓰기 위한 준비
with open('terminal_data.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # 헤더 작성
    writer.writerow(['터미널', '출발', '요금(어른)', '소요시간'])
"""
    # 각 터미널의 데이터를 추출
    for terminal in terminals:
        terminal_name = terminal.find_element(By.CSS_SELECTOR, '.toc').text  # 터미널 이름
        departure_times = terminal.find_elements(By.CSS_SELECTOR, '.departure')  # 출발 시간 목록
        prices = terminal.find_elements(By.CSS_SELECTOR, '.fare')  # 요금(어른)
        durations = terminal.find_elements(By.CSS_SELECTOR, '.duration')  # 소요 시간 목록

        # 출발 시간, 요금, 소요 시간 텍스트를 리스트로 저장
        for departure, price, duration in zip(departure_times, prices, durations):
            row = [
                terminal_name,
                departure.text,
                price.text.strip(),  # 요금에서 공백 제거
                duration.text.strip()  # 소요시간에서 공백 제거
            ]
            writer.writerow(row)

# 드라이버 종료
driver.quit()

print("데이터가 'terminal_data.csv'에 저장되었습니다.")
```
